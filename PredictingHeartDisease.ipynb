```python
#b.Check out the dataset. As you see, the dataset contains 
#a number of features including both contextual and biological factors (e.g. age, gender, vital signs, ...). 
#The last column “AHD” is the label with “Yes” meaning that a human subject has Heart Disease, and “No” meaning that 
#the subject does not have Heart Disease.
```

```python
#As you see, there are at least 3 categorical features in the dataset 
#(Gender, ChestPain, Thal). 
#Let’s ignore these categorical features for now, only keep the numerical features and 
#build your feature matrix and label vector.
```

```python
#### Step2: Set up the Feature Matrix and Label Vector:
```

```python
# Creating the Feature Matrix for heart dataset:

# create a python list of feature names that would like to pick from the dataset:
feature_cols = ['Age','RestBP','Chol','MaxHR','Oldpeak']

# use the above list to select the features from the original DataFrame
X = Heart_df[feature_cols] 

# print the first 5 rows
X.head()
```

```python
# checking the size of Feature Matix X:

print(X.shape)
```

```python
# select a Series of labels (the last column) from the DataFrame
y = Heart_df['ChestPain']

# checking the label vector by printing every 10 values
y[::10]
```

```python
#Step3: Defining (instantiating) an "object" from the sklearn class:
```

```python
# In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 

k = 3
knn = KNeighborsClassifier(n_neighbors=k) 
```

```python
# We can use the method "fit" of the "object knn" along with training dataset and labels to train the model.

knn.fit(X, y)
```

```python
# We can use the method "predict" of the *trained* object knn on one or more testing data sample to perform prediction:

X_Testing = [[63, 145, 233, 150, 2.3]]

y_predict = knn.predict(X_Testing)

print(y_predict)
```

```python
# We can use the method "predict" of the *trained* object knn on one or more testing data sample to perform prediction:
# Two new data samples:

X_Testing = [63, 145, 233, 150, 2.3],[41, 130, 204, 172, 1.4]

y_predict = knn.predict(X_Testing)

print(y_predict)
```

```python
# Evaluating the accuracy of our classifier:

#### 1- Let's split the Heart Dataset RANDOMLY into two new datasets: 
#Training Set (e.g. 70% of the dataset) and Testing Set (30% of the dataset).
#### 2- Let's pretend that we do NOT know the label of the Testing Set!
#### 3- Let's Train the model on only Training Set, and then Predict on the Testing Set!
#### 4- After prediction, we can compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy of our KNN Classifier!

#### We will learn more about model and accuracy evaluation in future tutorials!
```

```python
# Randomly splitting the original dataset into training set and testing set
# The function"train_test_split" from "sklearn.cross_validation" library performs random splitting.
# "test_size=0.3" means that pick 30% of data samples for testing set, and the rest (70%) for training set.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4)
```

```python
# print the size of the traning set:
print(X_train.shape)
print(y_train.shape)
```

```python
# print the size of the testing set:
print(X_test.shape)
print(y_test.shape)
```

```python
print(X_test)
print('\n')
print(y_test)
```

```python
#Training ONLY on the training set:
```

```python
knn.fit(X_train, y_train)
```

```python
#Testing on the testing set
```

```python
# Testing on the testing set:

y_predict = knn.predict(X_test)

print(y_predict)
```

```python
#Accuracy Evaluation:
#After prediction, we can now compare the "predicted labels" for the Testing Set with its "actual labels" 
#to evaluate the accuracy of our KNN Classifier!
```

```python
# We can now compare the "predicted labels" for the Testing Set with its "actual labels" to evaluate the accuracy 
# Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
# portion of correct predictions:

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)

print(accuracy)
```

```python
results = pd.DataFrame()

results['actual'] = y_test 
results['prediction'] = y_predict 

print(results)
```
